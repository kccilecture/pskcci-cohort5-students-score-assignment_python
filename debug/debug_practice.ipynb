```python
import random

def find_largest_number(start:int, end:int, numbers=5) -> int:
    random_list = [random.randint(start, end) for _ in range(numbers)]
    print("Random list: ", random_list)

    # Find the largest number.
    max_num = 0
    for num in random_list:
        if num > max_num:
            max_num = num
    print(f"Found largest number is {max_num}")
```

```python
find_largest_number(0, 100, 5)
```

```python
find_largest_number(-100, 0, 5)
```

Quiz: 0으로 나누기를 시도할 때 0.0을 반환하는 나눗셈 함수를 만들어 보세요

```python
def divide(a:int, b:int)-> float:
    return a / b
```

```python
assert divide(10, 2) == 5.0

import math
assert math.isclose(divide(10, 3), 3.333, rel_tol=1e-4)
```

```python
assert divide(10, 0) == 0.0
```

```python
import nltk
from nltk.tokenize import word_tokenize


def get_word_list(raw_text: str) -> list[str]:
    """문장부호를 제외한 문자열들을 골라서 리스트로 반환"""
    word_list = [word for word in word_tokenize(raw_text) if word.isalnum()]
    return word_list


def download_nltk_data():
    """ Download tokenizers if not exist."""
    try:
        nltk.data.find('tokenizers/punkt')
    except LookupError:
        nltk.download('punkt')

    try:
        nltk.data.find('tokenizers/punkt_tab')
    except LookupError:
        nltk.download('punkt_tab')
        
download_nltk_data()
```

```python
from collections import Counter

# 주홍색 연구
#text_file = "pg244.txt"
# KJV Bible
text_file = "bible_kjv.txt"
# 금강경
#text_file = "금강경.txt"

with open(text_file) as f:
    # 모든 단어들을 소문자로 변경
    text = f.read().lower()
    wl = get_word_list(text)

    # Counter를 이용한 단어 수 세기
    cnt = Counter(wl)

    # 전체 단어 카운트 수.
    total_word_cnt = sum(cnt.values())
    top_10 = cnt.most_common(10)

    # 화면 출력.
    print(f"Top 10 frequent words:")
    
    for i in range(10):
        k = top_10[i][0]
        v = top_10[i][1]
        avg = v / total_word_cnt
        print(f"\t{i+1}. {k} ({v} - Avg. {avg:.2f})")
```
