## 가변인자 (Variable Arguments)

```python
def print_장바구니(owner_id, owner_name, *args):
    print(f"{owner_name}({owner_id})님의 장바구니: ", end="")
    for a in args:
        print(f"{a} ", end="")
    print("")
```

```python

def personal_info(name, age, *args, **kwargs):
    print("-"*10)
    print(f"name: {name}")
    print(f"age: {age}")
    for k, v in kwargs.items():
        print(f"{k}: {v}")

    for a in args:
        print(f"{a}")        
```

## 재귀함수 (Recurrsion)

```python
def factorial(v):
    if v == 0:
        return 1
    return v * factorial(v - 1)
```

## Lambda 함수
    

```python
my_add_function = lambda a, b: a + b
print(my_add_function(10, 20))
```

```python
# 랜덤한 숫자 100개로 리스트 만들기
from random import randint
l = [randint(-100, 100) for x in range(100)]
print(l)
```

```python
# 모든 항목을 제곱한다.
m = map(lambda x: x**2, l)
print(list(m))
```

```python
students = [
  ('john', 'A', 200),
  ('jane', 'B', 12),
  ('dave', 'B', 100)]

print(students)
print(sorted(students, key=lambda x: x[1]))
```

## functools Module

```python
from functools import reduce, partial

r = reduce(lambda a, b: a + b, l)
print(r)
```

```python
def pow(base, exp):
    return base ** exp

square = partial(pow, exp=2)
print(square(5))
```
